```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_excel('./data.xlsx', sheet_name=None)
```

```python
df['Presents'].head()
```

```python
pdf = df['Presents'].dropna(how='all')
```

```python
rdf = df['Presents_2']
```

```python
replace_category = {}

for i, column in rdf.iteritems():
    # print(i,column)
    # name = column[0]
    category = column[1]
    items = []
    column = column.dropna()
    for j, v in enumerate(column[2:].array):
        items.append(v)
    replace_category[category] = items
```

```python
gift_raw = {}

for i, row in pdf.iterrows():
    # print(i)
    name = row[0]
    taste = row[1]
    items = []
    row = row.dropna()
    for j, v in enumerate(row[2:].array):
        if j % 2 == 0:
            items.append(v)
        else:
            items[j//2] = (items[j//2], v)
    gift_raw[(name, taste)] = items
```

```python
gift = []

for (name, taste), items in gift_raw.items():
    for (item, fp) in items:
        if item not in replace_category or len(replace_category[item]) == 0:
            gift.append([name, taste, item, fp])
            continue

        for new_item in replace_category[item]:
            gift.append([name, taste, new_item, fp])
```

```python
print(*gift, sep='\n')
```

```python
df['gift'] = pd.DataFrame(
    data=gift,
    columns=["name", "taste", "item", "fp"]
)
```

```python
writer = pd.ExcelWriter('newdata.xlsx')

for key, sheet in df.items():
    sheet.to_excel(writer, sheet_name=key, index=False)

writer.save()
```
