```python
import pandas as pd
```

```python
df = pd.read_excel('../xlsx/data.xlsx', sheet_name='Item Values',header=1)
df.head()
```

```python
items =  df.dropna(how='all',subset=['Item'])
```

```python
descs = []
with open('/home/ajohanes/Downloads/RF4TextDump/EngTxt/Item Descriptions.txt') as f:
    desc = ''
    multi = False

    while True:
        l = f.readline().strip().replace('"""','$"').replace('""',"$")

        if l == "":
            break

        if l[0] == '"' and not multi:
            multi = True
            desc = l[1:]
        elif l[-1] == '"' and multi:
            multi = False
            desc += " " + l[:-1]
            descs.append(desc.strip().replace('$','"'))
            desc = ''
        else:
            if multi:
                desc += " " + l
            else:
                desc = ''
                descs.append(l.replace('$','"'))
```

```python
data = []
for i, desc in enumerate(descs):
    item_name = items.iat[i,0].replace('&apos;',"'")
    item_desc = desc
    data.append([item_name,item_desc])

pd.DataFrame(
    data=data,
    columns=["name", "desc"]
).to_csv("../xlsx/item_desc.csv",index=False)
```
